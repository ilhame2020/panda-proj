```python
def load_data(file_path):
    if file_path.endswith('.xlsx'):
        return pd.read_excel(file_path)
    elif file_path.endswith('.json'):
        return pd.read_json(file_path)
    elif file_path.endswith('.csv'):
        return pd.read_csv(file_path, sep="|")
    else:
        raise ValueError("Unsupported file format")
```

```python
def columns_type(df):
    return df.dtypes
```

```python
#5.Count how many missing values there are in each column.
def count_missing_values(df):
    return df.isnull().sum()
```

```python
count_missing_values(data)
```

```python
#6.Which column or columns contain the highest number of missing values?
def column_with_most_missing_values(df):
    missing_counts = df.isnull().sum()
    max_missing = missing_counts.max()
    return missing_counts[missing_counts == max_missing].index.tolist()
```

```python
column_with_most_missing_values(data)
```

```python
#7.Remove all rows that contain any missing values and report how many rows remain.
def remove_rows_with_missing_values(df):
    cleaned_df = df.dropna()
    return cleaned_df.reset_index(drop=True)
```

```python
#8.Fill missing values in the city column with the string "Unknown".
def count_missing_city_with_unknown(df):
    df['city'] = df['city'].fillna("Unknown")

    return df.loc[:, ['city']].isnull().sum()
```

```python
#8.Fill missing values in the city column with the string "Unknown".
def fill_missing_city_with_unknown(df):
    df['city'] = df['city'].fillna("Unknown")
    return df
```

```python
count_missing_city_with_unknown(data)
```

```python
#9.Fill missing values in the unit_price column using the mean of that column.
def fill_missing_unit_price_with_mean(df):
    mean_price = df['unit_price'].mean()
    df['unit_price'] = df['unit_price'].fillna(mean_price)
    return df
```

```python
#10.Determine how many fully duplicated rows exist in the dataset.

def count_fully_duplicated_rows(df):
    return df.duplicated().sum()
```

```python
count_fully_duplicated_rows(data)
```

```python
#11.Remove duplicated rows and report the new number of rows.
def remove_duplicated_rows(df):
    cleaned_df = df.drop_duplicates()
    return cleaned_df, cleaned_df.shape[0]
```

```python
#12.Remove duplicates based only on the order_id column, keeping the first occurrence.
def remove_duplicates_based_on_order_id(df):
    cleaned_df = df.drop_duplicates(subset=['order_id'], keep='first')
    return cleaned_df, cleaned_df.shape[0]
```

```python
data=remove_duplicates_based_on_order_id(data)
```

```python
#global function data preparation takes only file path as input and executes all the above functions in sequence
def data_preparation(file_path):
    df = load_data(file_path)
    print("First 5 rows of the dataset:")
 
    print("\nColumn data types:")
    print(columns_type(df))
    print("\nMissing values in each column:")
    print(count_missing_values(df))
    print("\nColumn(s) with the most missing values:")
    print(column_with_most_missing_values(df))
    df_no_missing, rows_remaining = remove_rows_with_missing_values(df)
    print(f"\nNumber of rows after removing rows with missing values: {rows_remaining}")
    #fill the price free with 0 of unit_price
    df['unit_price'] = df['unit_price'].replace('free', 0)
    #check the unit_price values are not free anymore
    print("\nUnit price values after replacing 'free' with 0:")
    print(len(df[df['unit_price'] == 'free']))
    df_filled_city = fill_missing_city_with_unknown(df)

    duplicated_count = count_fully_duplicated_rows(df_filled_city)
    print(f"\nNumber of fully duplicated rows: {duplicated_count}")
    df_no_duplicates, new_row_count = remove_duplicated_rows(df_filled_city)
    print(f"\nNumber of rows after removing duplicated rows: {new_row_count}")
    df_final, final_row_count = remove_duplicates_based_on_order_id(df_no_duplicates)
    print(f"\nNumber of rows after removing duplicates based on order_id: {final_row_count}")
    return df_final
 
```
